#Predicting GPA Change from High School to College Using Machine Learning

```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('graduation_rate.csv')
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
highest_school_gpa = df.nlargest(10, 'high school gpa')
highest_school_gpa
```

```python
highest_college_gpa = df.nlargest(10, 'college gpa')
highest_college_gpa
```

```python
df['gpa_change'] = (df['college gpa'] > df['high school gpa'])
df['gpa_change']
```

```python
df['gpa_change'].value_counts()
```

```python
sns.countplot(x = 'gpa_change', data = df)
```
